## 4. Lists

Lists are ordered and changeable.


````python
numbers = [10, 20, 30, 40]

print(numbers[0])
print(numbers[-1])
print(numbers[1:3])

numbers.append(50)
numbers.remove(20)

print(numbers)
````

````python
# Useful list functions
values = [4, 2, 9, 1, 7]

print(len(values))
print(min(values))
print(max(values))
print(sum(values))
print(sorted(values))
````

---

## 5. Dictionaries

Dictionaries store key-value pairs.


````python
student = {
    "name": "Jan",
    "program": "MSDS",
    "status": "student"
}

print(student["name"])
print(student.get("program"))

student["course"] = "Machine Learning"

print(student.keys())
print(student.values())
print(student.items())
````

---

## 6. Conditionals

Use `if`, `elif`, and `else` to control logic.


````python
score = 87

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
else:
    grade = "Needs improvement"

print(grade)
````

---

## 7. Loops

Use loops to repeat actions.


````python
for number in range(5):
    print(number)
````

````python
names = ["Anna", "Ben", "Carlos"]

for name in names:
    print(name)
````

````python
for index, name in enumerate(names):
    print(index, name)
````

````python
courses = ["Python", "Statistics", "Machine Learning"]
grades = [95, 88, 91]

for course, grade in zip(courses, grades):
    print(course, grade)
````

## 8. Functions

Functions help avoid repeating code.


````python
def greet(name):
    return f"Hello, {name}."

print(greet("Annette"))
````

````python
def calculate_average(values):
    return sum(values) / len(values)

scores = [90, 85, 92, 88]
average = calculate_average(scores)

print(average)
````

---

## 9. List Comprehensions

List comprehensions are a shorter way to create lists.


````python
numbers = [1, 2, 3, 4, 5]

squared = [number ** 2 for number in numbers]

print(squared)
````

````python
even_numbers = [number for number in numbers if number % 2 == 0]

print(even_numbers)
````
