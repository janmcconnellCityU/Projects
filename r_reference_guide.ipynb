## 9. Data Manipulation with dplyr

The `dplyr` package provides functions for selecting, filtering, sorting, creating, and summarizing data.

The pipe operator `|>` passes the result of one operation to the next operation.

```r
# Install dplyr if needed
# install.packages("dplyr")

# Load dplyr
library(dplyr)

# Select columns
mtcars |>
    select(mpg, cyl, hp, wt)

# Filter rows
mtcars |>
    filter(mpg > 25)

# Sort rows
mtcars |>
    arrange(desc(mpg))

# Create a new column
mtcars |>
    mutate(power_to_weight = hp / wt)

# Combine multiple operations
mtcars |>
    filter(cyl == 6) |>
    select(mpg, hp, wt) |>
    arrange(desc(mpg))
```

## 10. Grouping and Summarizing

Grouped summaries calculate statistics for categories within a dataset.

The `group_by()` and `summarise()` functions are commonly used together to compare groups.

> **Tibble:** A modern version of an R data frame commonly returned by `dplyr` functions. Tibbles display column data types and print results in a more compact format.

```r
# Calculate summary statistics by cylinder count
mtcars |>
    group_by(cyl) |>
    summarise(
        vehicle_count = n(),
        mean_mpg = mean(mpg),
        mean_hp = mean(hp),
        mean_wt = mean(wt)
    )

# Group by multiple columns
mtcars |>
    group_by(cyl, gear) |>
    summarise(
        vehicle_count = n(),
        mean_mpg = mean(mpg),
        .groups = "drop"
    )
```

## 11. Joining Data Frames

Joins combine rows from two data frames using one or more matching columns.

Common joins include `left_join()`, `inner_join()`, `right_join()`, and `full_join()`.

```r
# Create example data frames
employees <- data.frame(
    employee_id = c(101, 102, 103, 104),
    employee_name = c("Alex", "Jordan", "Morgan", "Taylor"),
    department_id = c(1, 2, 1, 3)
)

departments <- data.frame(
    department_id = c(1, 2, 4),
    department_name = c("Finance", "Operations", "Analytics")
)

# Left join: keep all employees
employees |>
    left_join(
        departments,
        by = "department_id"
    )

# Inner join: keep matching rows only
employees |>
    inner_join(
        departments,
        by = "department_id"
    )

# Right join: keep all departments
employees |>
    right_join(
        departments,
        by = "department_id"
    )

# Full join: keep all rows from both data frames
employees |>
    full_join(
        departments,
        by = "department_id"
    )
```

## 14. Functions

Functions group reusable code into a named operation.

Arguments are defined inside `function()`, and the final evaluated expression is returned automatically unless `return()` is used explicitly.

```r
# Create a simple function
calculate_variance <- function(actual, expected) {
    actual - expected
}

# Call the function
calculate_variance(125, 100)

# Create a function with a default argument
calculate_change <- function(current, previous, percent = FALSE) {
    change <- current - previous

    if (percent) {
        return((change / previous) * 100)
    }

    change
}

# Calculate absolute change
calculate_change(125, 100)

# Calculate percent change
calculate_change(125, 100, percent = TRUE)
```

## 15. Conditional Logic

Conditional statements execute different code based on whether a logical condition is `TRUE` or `FALSE`.

Common structures include `if`, `else if`, and `else`.

```r
# Basic if statement
mpg_value <- 25

if (mpg_value > 20) {
    print("Above 20 MPG")
}

# If and else
if (mpg_value >= 30) {
    print("High fuel efficiency")
} else {
    print("Below high-efficiency threshold")
}

# Multiple conditions
if (mpg_value >= 30) {
    efficiency <- "High"
} else if (mpg_value >= 20) {
    efficiency <- "Moderate"
} else {
    efficiency <- "Low"
}

efficiency

# Vectorized conditional logic
mtcars |>
    mutate(
        efficiency = if_else(
            mpg >= 20,
            "Higher MPG",
            "Lower MPG"
        )
    ) |>
    select(mpg, efficiency) |>
    head()
```

## 16. Loops and Iteration

Loops repeat code across values or while a condition remains true.

R also provides vectorized and `apply` functions that can be more concise for many data analysis tasks.

**Note:** `apply()`, 1 means rows and 2 means columns.

```r
# For loop
for (value in c(10, 20, 30)) {
    print(value)
}

# While loop
count <- 1

while (count <= 3) {
    print(count)
    count <- count + 1
}

# Apply a function to each value
sapply(
    c(1, 2, 3, 4, 5),
    function(x) x ^ 2
)

# Apply a function to data frame columns
sapply(
    mtcars[c("mpg", "hp", "wt")],
    mean
)

# Row-level calculations
apply(
    mtcars[c("mpg", "hp", "wt")],
    1,
    mean
)
```

## 17. Data Visualization with ggplot2

The `ggplot2` package creates visualizations by combining data, aesthetic mappings, and geometric layers.

A basic plot begins with `ggplot()`, maps variables using `aes()`, and adds one or more `geom_` layers.

```r
# Install ggplot2 if needed
# install.packages("ggplot2")

# Load ggplot2
library(ggplot2)

# Scatter plot
ggplot(
    mtcars,
    aes(x = wt, y = mpg)
) +
    geom_point() +
    labs(
        title = "Fuel Efficiency by Vehicle Weight",
        x = "Weight (1,000 lbs)",
        y = "Miles per Gallon"
    )

# Scatter plot with a regression line
ggplot(
    mtcars,
    aes(x = wt, y = mpg)
) +
    geom_point() +
    geom_smooth(
        method = "lm",
        se = FALSE
    ) +
    labs(
        title = "Fuel Efficiency by Vehicle Weight",
        subtitle = "Linear regression trend",
        x = "Weight (1,000 lbs)",
        y = "Miles per Gallon"
    )

# Bar chart
ggplot(
    mtcars,
    aes(x = factor(cyl))
) +
    geom_bar() +
    labs(
        title = "Vehicles by Cylinder Count",
        x = "Cylinders",
        y = "Vehicle Count"
    )

# Histogram
ggplot(
    mtcars,
    aes(x = mpg)
) +
    geom_histogram(
        bins = 8
    ) +
    labs(
        title = "Distribution of Fuel Efficiency",
        x = "Miles per Gallon",
        y = "Count"
    )
```

## 18. Linear Regression

Linear regression models the relationship between a dependent variable and one or more explanatory variables.

The `lm()` function fits a linear model, and `summary()` provides coefficient estimates, statistical tests, and model fit statistics.

```r
# Fit a simple linear regression model
mpg_model <- lm(
    mpg ~ wt,
    data = mtcars
)

# View model coefficients
coef(mpg_model)

# View a detailed model summary
summary(mpg_model)

# Generate predicted values
predict(
    mpg_model,
    newdata = data.frame(
        wt = c(2.5, 3.0, 3.5)
    )
)

# Add fitted values and residuals
mtcars |>
    mutate(
        fitted_mpg = fitted(mpg_model),
        residual = residuals(mpg_model)
    ) |>
    select(mpg, wt, fitted_mpg, residual) |>
    head()
```

### Interpreting the Model Output

Key values from the regression summary include:

- **Intercept:** The model estimates an intercept of approximately `37.29`.
- **Weight coefficient:** The coefficient for `wt` is approximately `-5.34`. For each additional 1,000 pounds of vehicle weight, predicted fuel efficiency decreases by approximately 5.34 MPG.
- **p-value:** The very small p-value for `wt` indicates a statistically significant relationship between vehicle weight and MPG in this dataset.
- **R-squared:** The model explains approximately 75.3% of the variation in MPG.
- **Residual:** The difference between an observed value and the value predicted by the model. A negative residual means the observed MPG was lower than predicted; a positive residual means it was higher than predicted.

> **Note:** Statistical significance does not establish causation. The model describes the relationship between vehicle weight and MPG within the `mtcars` dataset.

## 19. Correlation

Correlation measures the strength and direction of a linear relationship between two numeric variables.

Correlation coefficients range from `-1` to `1`. Values closer to `-1` or `1` indicate stronger linear relationships, while values near `0` indicate weaker linear relationships.

```r
# Calculate correlation between two variables
cor(
    mtcars$wt,
    mtcars$mpg
)

# Calculate a correlation matrix
cor(
    mtcars[c("mpg", "hp", "wt", "qsec")]
)

# Calculate correlation using complete observations
cor(
    mtcars$wt,
    mtcars$mpg,
    use = "complete.obs"
)

# Perform a correlation test
cor.test(
    mtcars$wt,
    mtcars$mpg
)
```

### Interpreting Correlation Results

Key values from the correlation analysis include:

- **Correlation coefficient:** The correlation between vehicle weight and MPG is approximately `-0.868`, indicating a strong negative linear relationship. As vehicle weight increases, MPG generally decreases.
- **p-value:** The very small p-value provides strong evidence that the correlation is different from zero in this dataset.
- **95% confidence interval:** The estimated population correlation is between approximately `-0.934` and `-0.744`.
- **Correlation matrix:** The matrix compares pairwise correlations among multiple numeric variables. The diagonal values are `1` because each variable is perfectly correlated with itself.

> **Note:** Correlation measures association, not causation. A strong correlation does not establish that changes in one variable directly cause changes in another.

> **Regression connection:** In a simple linear regression with one explanatory variable and an intercept, the squared Pearson correlation is equal to the model's R-squared. Here, `(-0.868)² ≈ 0.753`, matching the regression model's R-squared.

### Comparing Excel and CSV Imports

The same product data can be interpreted differently depending on the source file and import function.

In the Excel worksheet, `Standard Cost` and `List Price` are stored as numeric values with currency formatting applied by Excel. `read_excel()` imports the underlying numeric values.

## 22. Cleaning and Converting Data Types

Imported data may require cleaning before analysis.

Character values containing currency symbols, thousands separators, or other formatting must be converted before they can be used in numeric calculations.

```r
# View the original column types
class(products_csv$Standard.Cost)
class(products_csv$List.Price)

# Remove currency formatting and convert to numeric
products_clean <- products_csv |>
    mutate(
        Standard.Cost = as.numeric(
            gsub(
                "[$,]",
                "",
                Standard.Cost
            )
        ),
        List.Price = as.numeric(
            gsub(
                "[$,]",
                "",
                List.Price
            )
        )
    )

# Check the converted column types
class(products_clean$Standard.Cost)
class(products_clean$List.Price)

# View the cleaned values
products_clean |>
    select(
        Product,
        Standard.Cost,
        List.Price
    ) |>
    head()

# Calculate summary statistics
products_clean |>
    summarise(
        mean_standard_cost = mean(Standard.Cost),
        mean_list_price = mean(List.Price)
    )
```

### Verifying Data Type Conversion

The currency-formatted CSV columns were initially imported as character data.

After removing dollar signs and thousands separators, `as.numeric()` successfully converted both columns to numeric values. The cleaned columns can now be used in calculations and statistical summaries.

> **Key point:** Confirm the resulting data type after conversion. A cleaning operation that appears to run successfully may still introduce `NA` values if unexpected text remains in the source data.

## 23. Duplicates and Distinct Values

Duplicate checks help identify repeated rows or values that may affect analysis.

The `duplicated()` function identifies duplicates, while `distinct()` returns unique rows or unique combinations of selected columns.

```r
# Count duplicated rows
sum(duplicated(products_clean))

# Identify duplicated product names
products_clean |>
    filter(
        duplicated(Product) |
        duplicated(Product, fromLast = TRUE)
    ) |>
    select(
        ProductKey,
        SKU,
        Product
    ) |>
    arrange(Product) |>
    head(10)

# Count distinct product names
products_clean |>
    summarise(
        row_count = n(),
        distinct_products = n_distinct(Product)
    )

# View distinct categories
products_clean |>
    distinct(Category) |>
    arrange(Category)

# View distinct category and subcategory combinations
products_clean |>
    distinct(
        Category,
        Subcategory
    ) |>
    arrange(
        Category,
        Subcategory
    )
```

### Interpreting Duplicate Checks

A duplicate value does not necessarily indicate a duplicate record.

The product dataset contains 397 rows and 295 distinct product names. Several product names appear more than once, but the associated records may have different product keys, SKUs, or other attributes.

The dataset also contains four distinct product categories and 37 category and subcategory combinations.

> **Key point:** Investigate duplicates using the columns that define a unique record. Removing rows based on a repeated name alone could discard valid data.

## 24. Renaming and Creating Columns

Columns can be renamed for clarity and new columns can be calculated from existing values.

The `rename()` function changes column names, while `mutate()` creates or modifies columns.

```r
# Rename columns
products_analysis <- products_clean |>
    rename(
        StandardCost = Standard.Cost,
        ListPrice = List.Price
    )

# Create calculated columns
products_analysis <- products_analysis |>
    mutate(
        PriceDifference = ListPrice - StandardCost,
        PriceDifferencePct = (
            PriceDifference / StandardCost
        ) * 100
    )

# View the calculated values
products_analysis |>
    select(
        Product,
        StandardCost,
        ListPrice,
        PriceDifference,
        PriceDifferencePct
    ) |>
    head()

# Summarize calculated columns
products_analysis |>
    summarise(
        mean_price_difference = mean(PriceDifference),
        median_price_difference = median(PriceDifference),
        mean_price_difference_pct = mean(PriceDifferencePct)
    )
```

### Validate the Percentage Calculation

Because `PriceDifferencePct` divides by `StandardCost`, zero values could produce undefined or infinite results.

Check for zero standard cost values before relying on the percentage calculation.

```r
products_analysis |>
    filter(StandardCost == 0) |>
    select(
        ProductKey,
        Product,
        StandardCost,
        ListPrice
    )
```

## 25. String Manipulation

Character data often requires cleaning, searching, or extracting text.

Base R provides functions for common string operations.

```r
# Convert text to uppercase and lowercase
toupper(products_analysis$Category[1:5])
tolower(products_analysis$Category[1:5])

# Count characters
nchar(products_analysis$Product[1:5])

# Search for text
grepl(
    "Helmet",
    products_analysis$Product[1:10]
)

# Filter rows containing text
products_analysis |>
    filter(
        grepl("Helmet", Product)
    ) |>
    select(
        ProductKey,
        Product,
        Category,
        Subcategory
    ) |>
    head()

# Replace text
gsub(
    "Helmet",
    "HELMET",
    products_analysis$Product[1:5]
)

# Extract part of a string
substr(
    products_analysis$SKU[1:5],
    1,
    2
)
```

### String Function Behavior

Functions such as `toupper()`, `tolower()`, `gsub()`, and `substr()` return modified values but do not change the original data unless the result is assigned to an object or column.

`grepl()` returns `TRUE` or `FALSE` for each value and is useful for filtering rows based on text patterns.

> **Key point:** R string searches are case-sensitive by default. `"Helmet"` and `"helmet"` are treated as different text unless case handling is specified.

### Using stringr

The `stringr` package provides consistent functions for common string operations and is frequently used with tidyverse workflows.

Many `stringr` functions provide equivalents to base R string functions with standardized `str_` names.

```r
# Install stringr if needed
# install.packages("stringr")

# Load stringr
library(stringr)

# Convert text to uppercase and lowercase
str_to_upper(products_analysis$Category[1:5])
str_to_lower(products_analysis$Category[1:5])

# Count characters
str_length(products_analysis$Product[1:5])

# Detect text
str_detect(
    products_analysis$Product[1:10],
    "Helmet"
)

# Filter rows containing text
products_analysis |>
    filter(
        str_detect(Product, "Helmet")
    ) |>
    select(
        ProductKey,
        Product,
        Category,
        Subcategory
    ) |>
    head()

# Replace text
str_replace_all(
    products_analysis$Product[1:5],
    "Helmet",
    "HELMET"
)

# Extract the first two characters
str_sub(
    products_analysis$SKU[1:5],
    1,
    2
)
```

### Base R and stringr Comparison

| Base R | `stringr` |
|---|---|
| `toupper()` | `str_to_upper()` |
| `tolower()` | `str_to_lower()` |
| `nchar()` | `str_length()` |
| `grepl()` | `str_detect()` |
| `gsub()` | `str_replace_all()` |
| `substr()` | `str_sub()` |

Both approaches perform similar string operations. The `stringr` functions use consistent `str_` naming and integrate naturally with tidyverse workflows.

## 26. Managing Missing Values in Data Frames

R represents missing values with `NA`.

Missing values can affect calculations, filtering, and statistical functions and should be identified and handled deliberately.

```r
# Create example data with missing values
project_status <- data.frame(
    project = c(
        "Alpha",
        "Beta",
        "Gamma",
        "Delta",
        "Epsilon"
    ),
    budget = c(
        125000,
        98000,
        NA,
        150000,
        110000
    ),
    completion_pct = c(
        100,
        75,
        60,
        NA,
        90
    )
)

# View the data
project_status

# Identify missing values
is.na(project_status)

# Count missing values by column
colSums(
    is.na(project_status)
)

# Calculate a mean with missing values
mean(project_status$budget)

# Ignore missing values in the calculation
mean(
    project_status$budget,
    na.rm = TRUE
)

# Return rows containing a missing value
project_status |>
    filter(
        if_any(
            everything(),
            is.na
        )
    )

# Remove rows containing missing values
project_status |>
    drop_na()

# Replace missing values
project_status |>
    replace_na(
        list(
            budget = 0,
            completion_pct = 0
        )
    )
```

### Handling Missing Values Deliberately

Many R functions return `NA` when the input contains missing values. The `na.rm = TRUE` argument tells supported functions to exclude missing values from the calculation.

Removing or replacing missing values changes the data and should reflect the analytical context. Replacing an unknown budget or completion percentage with `0`, for example, is not automatically correct.

> **Key point:** `NA` means a value is missing or unknown. It does not mean zero. Inspect why values are missing before removing or replacing them.

## 27. Sorting and Ranking Data

Rows can be sorted by one or more columns, and observations can be ranked based on numeric values.

The `arrange()`, `slice_max()`, and `slice_min()` functions are useful for identifying high and low values in a dataset.

```r
# Sort products by list price
products_analysis |>
    arrange(ListPrice) |>
    select(
        Product,
        ListPrice
    ) |>
    head()

# Sort in descending order
products_analysis |>
    arrange(desc(ListPrice)) |>
    select(
        Product,
        ListPrice
    ) |>
    head()

# Return the five highest list prices
products_analysis |>
    slice_max(
        order_by = ListPrice,
        n = 5
    ) |>
    select(
        Product,
        ListPrice
    )

# Return the five lowest list prices
products_analysis |>
    slice_min(
        order_by = ListPrice,
        n = 5
    ) |>
    select(
        Product,
        ListPrice
    )

# Create a price rank
products_analysis |>
    mutate(
        PriceRank = min_rank(
            desc(ListPrice)
        )
    ) |>
    select(
        Product,
        ListPrice,
        PriceRank
    ) |>
    arrange(PriceRank) |>
    head(10)
```

### Understanding Ties in Rankings

Equal values can affect both row selection and ranking results.

The five `Road-150 Red` products share the highest list price, so `min_rank()` assigns each product a rank of `1`. The next highest price receives rank `6` because five observations occupy the first ranking position.

`slice_max()` and `slice_min()` include tied values at the cutoff by default and may therefore return more rows than the value specified by `n`.

> **Key point:** Ranking functions handle ties differently. Confirm the desired ranking behavior before using ranks in reporting or analysis.

## 28. Grouped Calculations with Multiple Statistics

Grouped summaries calculate statistics separately for categories or other groups.

Multiple summary measures can be calculated in a single `summarise()` operation.

```r
# Calculate multiple statistics by category
category_summary <- products_analysis |>
    group_by(Category) |>
    summarise(
        product_count = n(),
        mean_list_price = mean(ListPrice),
        median_list_price = median(ListPrice),
        min_list_price = min(ListPrice),
        max_list_price = max(ListPrice),
        mean_standard_cost = mean(StandardCost)
    )

# View the grouped summary
category_summary

# Calculate statistics across multiple columns
products_analysis |>
    group_by(Category) |>
    summarise(
        across(
            c(StandardCost, ListPrice),
            list(
                mean = mean,
                median = median,
                min = min,
                max = max
            )
        )
    )

# Group by multiple columns
products_analysis |>
    group_by(
        Category,
        Subcategory
    ) |>
    summarise(
        product_count = n(),
        mean_list_price = mean(ListPrice),
        .groups = "drop"
    ) |>
    arrange(
        Category,
        desc(mean_list_price)
    )
```

### Interpreting Grouped Summaries

Grouped summaries calculate statistics within defined categories rather than across the entire dataset.

The AdventureWorks product categories have substantially different price ranges. For example, Bikes have a mean list price of approximately `$1,524.59`, while Accessories average approximately `$34.23`.

The `across()` function applies the same set of summary functions to multiple columns and automatically creates names such as `StandardCost_mean` and `ListPrice_median`.

> **Key point:** Overall summary statistics can hide meaningful differences between groups. Examine grouped results when categories represent distinct products, populations, or business segments.

## 29. Statistical Tests

Statistical tests evaluate evidence about differences, relationships, or other characteristics in sample data.

The `t.test()` function can compare the means of two groups.

```r
# Filter two product categories
bike_component_prices <- products_analysis |>
    filter(
        Category %in% c(
            "Bikes",
            "Components"
        )
    )

# Compare group summary statistics
bike_component_prices |>
    group_by(Category) |>
    summarise(
        product_count = n(),
        mean_list_price = mean(ListPrice),
        median_list_price = median(ListPrice),
        standard_deviation = sd(ListPrice)
    )

# Perform a two-sample t-test
price_test <- t.test(
    ListPrice ~ Category,
    data = bike_component_prices
)

# View the test results
price_test
```

### Interpreting the t-Test

The mean list price for Bikes is approximately `$1,524.59`, compared with approximately `$551.11` for Components.

The Welch two-sample t-test reports a very small p-value (`p < 2.2e-16`), providing strong evidence that the mean list prices differ between the two categories in this dataset.

The 95% confidence interval estimates that the difference in mean list price is between approximately `$793.64` and `$1,153.33`.

> **Key point:** Statistical significance indicates evidence of a difference between group means. It does not explain why the groups differ or establish a causal relationship.

> **Note:** `t.test()` uses Welch's two-sample t-test by default when comparing two groups. Welch's test does not assume equal population variances.
